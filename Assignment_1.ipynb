### BFS

```python
def bfs(graph, start, goal):
    visited = []
    queue = deque([start])
    steps = 0

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.append(node)
            steps += 1
            if node == goal:
                break
            queue.extend([neighbor for neighbor, _ in graph[node] if neighbor not in visited])

    return visited, steps

# loop all test cases
for test_case_num in range(1, 4):
    edge_list, node_ids, start_node, goal_node = load_files_for_test_case(test_case_num)

    start_time = time.time()
    bfs_path, steps_taken = bfs(edge_list, start_node, goal_node)
    end_time = time.time()

    print(f"\nTest Case {test_case_num}:")
    print(f"BFS Path: {bfs_path}")
    print(f"Total Steps Taken: {steps_taken}")
    print(f"BFS Execution Time: {end_time - start_time:.4f} seconds")
```

### DFS

```python
def dfs(graph, start, goal):
    visited = []
    stack = [start]
    steps = 0

    while stack:
        node = stack.pop()
        if node not in visited:
            visited.append(node)
            steps += 1
            if node == goal:
                break
            stack.extend([neighbor for neighbor, _ in graph[node] if neighbor not in visited])

    return visited, steps

# Loop all test cases
for test_case_num in range(1, 4):
    edge_list, node_ids, start_node, goal_node = load_files_for_test_case(test_case_num)

    start_time = time.time()
    dfs_path, steps_taken = dfs(edge_list, start_node, goal_node)
    end_time = time.time()

    print(f"\nTest Case {test_case_num}:")
    print(f"DFS Path: {dfs_path}")
    print(f"Total Steps Taken: {steps_taken}")
    print(f"DFS Execution Time: {end_time - start_time:.4f} seconds")
```

### A* with 2 Heuristics

```python
# 2 Heuristic functions
def manhattan_distance(coord1, coord2):
    return abs(coord1[0] - coord2[0]) + abs(coord1[1] - coord2[1])

def euclidean_distance(coord1, coord2):
    return math.sqrt((coord1[0] - coord2[0])**2 + (coord1[1] - coord2[1])**2)

# A* search function
def astar(graph, coordinates, start, goal, heuristic):
    open_set = []
    heapq.heappush(open_set, (0, start))
    came_from = {}
    g_score = {node: float('inf') for node in graph}
    g_score[start] = 0
    f_score = {node: float('inf') for node in graph}
    f_score[start] = heuristic(coordinates[start], coordinates[goal])

    steps = 0

    while open_set:
        current = heapq.heappop(open_set)[1]
        steps += 1
        if current == goal:
            break
        for neighbor, weight in graph[current]:
            tentative_g_score = g_score[current] + weight
            if tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] =tentative_g_score
                f_score[neighbor]= g_score[neighbor] +heuristic(coordinates[neighbor],coordinates[goal])
                heapq.heappush(open_set, (f_score[neighbor], neighbor))

    return reconstruct_path(came_from, start, goal), steps

# reconstruction function
def reconstruct_path(came_from, start, goal):
    current = goal
    path = [current]
    while current in came_from:
        current = came_from[current]
        path.append(current)
    path.reverse()
    return path

# function for loops with both hueristics
def run_astar_for_test_case(test_case_num):
    edge_list, node_ids, start_node, goal_node = load_files_for_test_case(test_case_num)

    print(f"\nTest Case {test_case_num}:")
    start_time = time.time()
    astar_manhattan_path, manhattan_steps = astar(edge_list, node_ids, start_node, goal_node, manhattan_distance)
    end_time = time.time()
    print(f"A* with Manhattan Heuristic Path (Test Case {test_case_num}): {astar_manhattan_path}")
    print(f"A* with Manhattan Heuristic took {end_time - start_time:.4f} seconds")
    print(f"Total Steps Taken with Manhattan Heuristic: {manhattan_steps}\n")

    start_time = time.time()
    astar_euclidean_path, euclidean_steps = astar(edge_list, node_ids, start_node, goal_node, euclidean_distance)
    end_time = time.time()
    print(f"A* with Euclidean Heuristic Path (Test Case {test_case_num}): {astar_euclidean_path}")
    print(f"A* with Euclidean Heuristic took {end_time - start_time:.4f} seconds")
    print(f"Total Steps Taken with Euclidean Heuristic: {euclidean_steps}\n")

# loop for all test cases
for test_case_num in range(1, 4):
    run_astar_for_test_case(test_case_num)
```

The goal of this assignment is to design a robot that can move from a start location to a goal using search algorithms. The robot needs to evaluate the ability of both uninformed and informed searches to find the optimal path between two locations. Three algorithms were implemented:

## Heuristic Choices and Performance

For A*, two heuristics were used:
- **Manhattan Heuristic**: Measures the sum of the absolute x and y coordinate differences.
- **Euclidean Heuristic**: Measures the straight-line distance.

### Admissibility of Heuristics

In the test cases, movement is restricted to horizontal and vertical paths. This makes the **Manhattan heuristic** admissible, as it accurately calculates the cost based on the sum of those distances. It never underestimates the true cost.

The **Euclidean heuristic** is also admissible because, although diagonal movement does not occur, it still provides a lower bound that never overestimates the true path cost. The correctness of both heuristics being admissible is further shown by their performance compared to BFS and DFS with A star consistently founding optimal path without overestimating the cost, while the uninformed searches took more steps and less efficient routes in the tests.
