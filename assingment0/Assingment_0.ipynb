```python
import numpy as np
import matplotlib.pyplot as plt

```

### Problem 1

```python

array = np.random.uniform(low = 0, high =5, size = 5)
print("Starting array:"  , array, "\n")

mean = np.mean(array)
std = np.std(array)
print(f"Mean: {mean} \nStandard Deviation: {std} \n")

reshaped_array = array.reshape(5, 1)
print("Reshaped array:\n" , reshaped_array, "\n")

array_5_more = reshaped_array + 5
print("Array 5 added:\n" , array_5_more, "\n")

dot_product = np.dot(array_5_more.T, array_5_more)
print("Dot Product with itself:\n", dot_product)
```

### Problem 2

```python
x = np.arange(0,100, .1)
print("x =", x[0], "through", x[-1], "\n")

y = np.sin(x)
print("y =", y[0], "through", y[-1], "\n")

plt.plot(x,y)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Sine Wave")

plt.savefig("sine_wave.png")
```

### Problem 3

```python
x = np.arange(0, 101, 1)
print("x =", x[0], "through", x[-1], "\n")

y = .5 * x**2 +2 * x +1
print("y =", y[0], "through", y[-1], "\n")

plt.plot(x, y, label="y = 0.5x^2 + 2x + 1")
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Quadratic Function")
plt.legend()
plt.grid(True, linestyle=":")

plt.savefig("quadratic_function.pdf")
```
